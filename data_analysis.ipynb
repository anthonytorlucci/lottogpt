### verify the data honors the constraints (if any)
In this case, each sequence should be 3 number and they should be sorted by increasing value.
Also, the minimum and maximum values for each number in the sequence are:
- first number : [1, 31]
- second number : [2, 32]
- third number : [3, 33]

```python
from data.data import lottery_data
ldat = lottery_data()
for seq in ldat:
    assert len(seq) == 3
    assert seq == sorted(seq)
    assert seq[0] > 0 and seq[0] < 32
    assert seq[1] > 1 and seq[1] < 33
    assert seq[2] > 2 and seq[2] < 34
```

```python
import numpy
import seaborn as sns
sns.set_style("dark")

seq0 = numpy.array([x[0] for x in ldat])
print(f'min:{numpy.min(seq0)}, max:{numpy.max(seq0)}')
counts, bin_edges = numpy.histogram(seq0, bins=int(numpy.max(seq0))-int(numpy.min(seq0)))
print(bin_edges)
print(counts)
sns.displot(seq0, binwidth=1)
```

Most people might expect random drawings to yield a "white" spectrum where each number drawn is equally likely. However, these numbers are sorted and they are sorted *after* they have been drawn. Only the first 3 of 5 numbers are being used and that skews the distribution.

```python
seq1 = numpy.array([x[1] for x in ldat])
print(f'min:{numpy.min(seq1)}, max:{numpy.max(seq1)}')
counts, bin_edges = numpy.histogram(seq1, bins=int(numpy.max(seq1))-int(numpy.min(seq1)))
print(bin_edges)
print(counts)
sns.displot(seq1, binwidth=1)
```
